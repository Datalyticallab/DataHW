````python
from math import *

def calc_area(r):       # правки в назві - усі букви маленькі, між словами нижнє підкреслення
    return pi * r ** 2 # перенесла на окремий рядок, додала пробіли біля операторів, зробила відступ, додала рядок між функціями

def calc_circumference(r):   # правки в назві - усі букви маленькі, між словами нижнє підкреслення
    return 2 * pi * r     # перенесла на окремий рядок, додала пробіли біля операторів, зробила відступ

radius = 5
print("Area:", calc_area(radius), "Circumference:", calc_circumference(radius)) # пдодала пробіли
````

````python
def CalculateVariance(nl):
    sum_list = 0
    for n in nl:
        sum_list = sum_list + n
    mean = sum_list / len(nl)
    sum_squares = 0
    for n in nl:
        sum_squares = sum_squares + n**2
    mean_squares = sum_squares / len(nl)
    return mean_squares
````

````python
input_list = [75, 80, 85, 90, 95]
````

````python
def calculate_variance(num_list):
    #calculate the mean
    mean = sum(num_list) / len(num_list)

    #calculate the sum of squared deviations from the mean
    sum_squared_deviation = sum((n - mean) ** 2 for n in num_list)

    #calculate the variance
    variance = sum_squared_deviation / len(num_list)

    return variance

calculate_variance(input_list)
````
